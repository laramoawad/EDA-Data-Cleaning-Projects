# Data Cleaning in Pandas 

```python
import pandas as pd
```

```python
df = pd.read_excel(r"C:\Users\Asus\Downloads\Customer Call List.xlsx")
df

#need to clean this up and send it to sales people to make calls quick as possible and all info standardized and get rid of not useful columns and rows
#that say yes do not contact 
```

```python
#drop 1 of row 19 and 20 cuz same 
df = df.drop_duplicates()
df
```

```python
#remove un-needed columns 
#to save this change u need to add it to df = [action u want]
df = df.drop(columns = 'Not_Useful_Column')
df

#or

#del df['Not_Useful_Column']
#df
```

```python
#need to clean up last name cuz fee 7agaat not consistent 
#need to specify the column u want to fix 
#strip used if the things u want removing are either at beginning or end not in middle of wording
#df['Last_Name']= df['Last_Name'].str.lstrip('...') #removed dots from Potter 
#df['Last_Name']=df['Last_Name'].str.lstrip('/') #removed / from White 
#now fix right side so rstrip
#df['Last_Name']=df['Last_Name'].str.rstrip('_') #removed / from White 
#u can do the above bas actually strip works differently, need to put all in one code 

#I know i wanna remove the dot and the slash and the underscore and law fee any numbers remove them tooo..so 

df['Last_Name']= df['Last_Name'].str.strip('123./_')
df
```

```python
#workign with phone numbers is a bit hard and there are so many formats & u need somethign to automate that and fix it
#we want all phone numbers to look like first one 
```

```python
#regualr expresion

#str.replace('[^a-zA-Z0-9]', '', regex=True): 
      #This pattern removes any character that is NOT a letter (a-z, A-Z) or a digit (0-9). 
      #^: Inside the square brackets, the caret (^) acts as a negation. It means "anything not in this set."
#regex=True: This ensures that pandas treats the pattern as a regular expression. 
      #In newer versions of pandas, regex=True is required to make the function work as expected.

df['Phone_Number']= df['Phone_Number'].str.replace('[^a-zA-Z0-9]', '', regex=True)
df
```

```python
#it doesnt include the 3 it goes up to it so first char is 0 then 1 then 2 bas u need to include up to 3 for it to include 2
#df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10]) #here we are looking at as if its a string but the dtype is a float so hayitla3 error 

#we need to change column into a string before we apply the formatting 


#run this first to consider all as string
df['Phone_Number']=df['Phone_Number'].apply(lambda x: str(x))

#now run this again and it will work 
df['Phone_Number']=df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
df
```

```python
#In summary, strip() is for trimming characters from the edges of strings, 
#while replace() is for replacing specific characters or patterns within strings so cannot use string below only replace

df['Phone_Number']=df['Phone_Number'].str.replace('nan--','')
df['Phone_Number']=df['Phone_Number'].str.replace('Na--','')
df
```

```python
# we will look at address now and want to seperate it into 3 columns and see column for zip code w kida 
df[['Street_Address', 'State', 'Zip_Code']] = df['Address'].str.split(',',n=2, expand = True)
df

```

```python
#now standardize all yes and no as Y and N only 
df['Paying Customer']= df['Paying Customer'].str.replace('Yes', 'Y')
df['Paying Customer']=df['Paying Customer'].str.replace('No', 'N')
df
```

```python
df['Do_Not_Contact']= df['Do_Not_Contact'].str.replace('Yes', 'Y')
df['Do_Not_Contact']=df['Do_Not_Contact'].str.replace('No', 'N')
df
```

```python
#remove str from here cuz ur just looking across whole df not specific column as before where u mentioned the column name and str
df= df.replace('N/a', ' ') #pyhton mish faahim n/a is same as none and nan so need to do it seperately 

df = df.fillna('') #im saying anything that is a nan or none leave it blank 
```

```python
#we cleaned: we removed columns, split col, formatted and cleaned phone numbers, too values away from last_name col, standardized paying customer and do not contact too

df
```

```python
#we cannot give ther people in call center the numbers of people that have a yes in do not contact col 
#and dont give them names of people with no phone numbers aslan 
#remove those 

#if Y then dont call 

#do 4 loop 

#use loc ot look at indexes and im specifying which column to look at 

for x in df.index:
    if df.loc[x, 'Do_Not_Contact'] == 'Y':
        df.drop(x, inplace = True)

df

#so now kida 1 and 3 and 18 are missing and there are no more Y in Do_Not_Contact column
```

```python
#also if phone number blank i dont want to send these rows to the call center people

for x in df.index:
    if df.loc[x, 'Phone_Number'] == '':
        df.drop(x, inplace = True)

df

#or another way to drop null values: 
#df = df.dropna(subset=['Phone_Number'], inplace = True)




#now a lot of rows are removed and only those with phone numbers and Ns are there which the call center should call 
```
